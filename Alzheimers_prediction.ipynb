#outliers

```python
import numpy as np

numerical_cols = ["Age", "Education Level", "BMI", "Cognitive Test Score"]

outliers = {}
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
{col: len(outliers[col]) for col in outliers}
```

```python
import pandas as pd
import matplotlib.pyplot as plt


features = ["Physical Activity Level", "Depression Level", "Air Pollution Exposure", "Stress Levels"]


countries_filtered = df[df["Country"].isin(["Japan", "Russia"])]


percentage_df = (
    countries_filtered.groupby("Country")[features]
    .apply(lambda x: x.apply(lambda col: col.value_counts(normalize=True).mul(100)))  #col.value_counts(normalize=True)=Now, instead of raw counts, it returns proportions
)                                                                                     #.mul(100)  multiplying by 100 converts these proportions into percentages:


colors = {"Japan": "#1f77b4", "Russia": "#d62728"}

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, feature in enumerate(features):
    category_data = percentage_df[feature].unstack(level=0)


    category_data = category_data.T
    category_data.index.name = "Category"
    category_data.reset_index(inplace=True)


    category_data.set_index("Category").plot(
        kind="bar", stacked=True, ax=axes[i], color=[colors["Japan"], colors["Russia"]]
    )

    axes[i].set_title(f"{feature} Distribution: Japan vs. Russia", fontsize=12)
    axes[i].set_ylabel("Percentage")
    axes[i].set_xlabel("Category")


    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=45, ha="right", fontsize=10)


    axes[i].grid(axis="y", linestyle="--", alpha=0.7)
    axes[i].legend(title="Country", loc="upper right", bbox_to_anchor=(1.15, 1))

plt.tight_layout()
plt.show()
```

# train test split

```python
x=df.drop('Alzheimer’s Diagnosis',axis=1)
y=df['Alzheimer’s Diagnosis']
```

```python
from sklearn.preprocessing import StandardScaler
sd=StandardScaler()
sd.fit(x)
x=sd.transform(x)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state =42,stratify=y)
```

```python
ytrain.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE()
xres,yres=sm.fit_resample(xtrain,ytrain)
```

```python
yres.value_counts()
```

#KNN

BEFORE SMOTING

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

knn=KNeighborsClassifier()
knn.fit(xtrain,ytrain)

ypred=knn.predict(xtest)

acc=accuracy_score(ytest,ypred)
print(acc)
cn=confusion_matrix(ytest,ypred)
print(cn)
cr=classification_report(ytest,ypred)
print(cr)
```

```python
training_score=knn.score(xtrain,ytrain)
print(training_score)
```

```python
testing_score=knn.score(xtest,ytest)
print(testing_score)
```

#Final (XGBOOST)

XGBoost (Extreme Gradient Boosting) is an advanced ensemble learning algorithm that:

 Builds multiple decision trees sequentially, improving performance at each step.

 Finds non-linear relationships that correlation does not detect.

 Combines weak signals from multiple features to make strong predictions.

Instead of relying on one strong feature, XGBoost learns patterns across multiple weak features using an iterative approach

after hyperparameter tuning

```python
from sklearn.metrics import ConfusionMatrixDisplay

plt.figure(figsize=(6,4))
sns.heatmap(cn_hy, annot=True, fmt='d', cmap='Blues', xticklabels=['No AD', 'AD'], yticklabels=['No AD', 'AD'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

True Positives (AD correctly predicted as AD): 4,521

True Negatives (No AD correctly predicted as No AD): 6,318

False Positives (No AD incorrectly predicted as AD): 2,396

False Negatives (AD incorrectly predicted as No AD): 1,622

The confusion matrix shows that there are 2,396 false positives (FP)—cases where the model incorrectly predicted AD for someone who does not have it—and 1,622 false negatives (FN)—cases where the model incorrectly predicted No AD for someone who actually has it.

Since FP (2,396) > FN (1,622), the model is slightly more likely to misclassify a person without AD as having AD than to miss an actual AD case. This means the model prioritizes sensitivity (catching true AD cases) more than specificity (avoiding false alarms)

```python
y_pred_prob = xgb_hy.predict_proba(xtest)[:, 1]  # Get probability of the positive class

# Compute ROC curve
fpr, tpr, _ = roc_curve(ytest, y_pred_prob)
roc_auc = auc(fpr, tpr)

# Plot ROC Curve
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Diagonal reference line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

Your curve is above the dashed diagonal line, meaning your model performs better than random guessing.

The AUC (Area Under the Curve) is 0.80, which means your model has good discriminatory ability (it correctly distinguishes between AD and No AD cases 80% of the time).

```python
precision, recall, _ = precision_recall_curve(ytest, y_pred_prob)

plt.figure(figsize=(6,4))
plt.plot(recall, precision, color='blue', lw=2)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.show()
```

The curve starts at (0,1), meaning at very low recall, precision is high.
As recall increases, precision decreases, showing the trade-off.

At high recall (close to 1.0), precision drops to around 0.4, meaning the model correctly identifies most AD cases but also makes many false positive errors.

At High Recall (~1.0),

Precision is Low (~0.4):

The model correctly identifies almost all AD cases.
But many of the predicted AD cases are actually wrong → too many false positives.
